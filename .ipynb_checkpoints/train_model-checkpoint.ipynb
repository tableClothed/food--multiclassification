```python
import tensorflow as tf
from tensorflow import keras
import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/data.csv")

categories = list(df['item'])
# categories
```

```python
X, y = [], []
for label in categories:  
    path = os.path.join('data\\images', label)
    images_list = []
    for img in os.listdir(path):
        image = os.path.join(path, img)
        image = cv2.imread(image, 0)
        image_res = cv2.resize(image, (224, 224))
        X.append(image_res)
        y.append(label)
```

```python
plt.imshow(X[1], cmap="gray")
plt.show()
```

```python
X = np.array(X)
X = X/255.0
y = np.array(y)
```

```python
from tensorflow.keras.applications import ResNet50, MobileNet, Xception, DenseNet121
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Input, Dense

img = Input(shape=(224, 224, 3))
```

```python
# RESNET50
model = DenseNet121(include_top=False, weights='imagenet', input_tensor=img,
                  input_shape=None, pooling='avg')

final = model.layers[-1].output
dense = Dense(128, activation="relu")(final)
output = Dense(10, activation="sigmoid")(dense)
model = Model(img, output)

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=['accuracy'])
model.save('densenet_model.h5')
```

```python
# MOBILENET
model = MobileNet(include_top=False, weights='imagenet', input_tensor=img,
                  input_shape=None, pooling='avg')

final = model.layers[-1].output
dense = Dense(128, activation="relu")(final)
output = Dense(10, activation="sigmoid")(dense)
model = Model(img, output)

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=['accuracy'])
model.save('mobilenet_model.h5')
```

```python
# XCEPTION
model = Xception(include_top=False, weights='imagenet', input_tensor=img,
                  input_shape=None, pooling='avg')

final = model.layers[-1].output
dense = Dense(128, activation="relu")(final)
output = Dense(10, activation="sigmoid")(dense)
model = Model(img, output)

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=['accuracy'])
model.save('xception_model.h5')
```

```python
# RESNET
model = ResNet50(include_top=False, weights='imagenet', input_tensor=img,
                  input_shape=None, pooling='avg')

final = model.layers[-1].output
dense = Dense(128, activation="relu")(final)
output = Dense(10, activation="sigmoid")(dense)
model = Model(img, output)

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=['accuracy'])
model.save('resnet_model.h5')
```

```python
# SEQUENTIAL
model = Sequential()
model.add(Dense(128, activation="relu"))
model.add(Dense(10, activation="sigmoid"))

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=['accuracy'])
model.save('sequential_model.h5')
```
